Импортируем необходимые библеотеки.

```python
import pandas as pd
import numpy as np
```

Загружаем датасет для обучения.

```python
train = pd.read_csv("titanic_train.csv")
```

```python
train
```

Посчитаем количество пустых значений в выборке.

```python
train.isnull().sum()
```

Заменим значения в полях возраста, там где значения отсутсвуют, на средний возраст.

```python
ages = train['Age']
ages
```

```python
sum = 0
for i in range(0, len(ages)):
    if (ages[i] >= 0):
        sum += ages[i]

average = sum/(len(ages)-ages.isnull().sum())
average
```

```python
train['Age'] = train['Age'].fillna(average)
train
```

```python
train.isnull().sum()
```

Заменим значения 'Embarked', там, где оно отсутствует, на самое частое значение в столбце. 

```python
train['Embarked'].value_counts().idxmax()
```

```python
train['Embarked'] = train['Embarked'].fillna('S')

for i in range(0, len(train['Embarked'])):
    if (train['Embarked'][i] == 'C'):
        train['Embarked'][i] = 0
    if (train['Embarked'][i] == 'Q'):
        train['Embarked'][i] = 1
    if (train['Embarked'][i] == 'S'):
        train['Embarked'][i] = 2

train.isnull().sum()
```

У 77% пассажиров не указано значение в графе 'Cabin'. Будет разумно не учитывать этот столбец вовсе, вместо заполнения NaN.

Заменим значения графе 'Sex' на '1' и '0', где '1' - мужчина.

```python
for i in range(0, len(train['Sex'])):
    if (train['Sex'][i] == 'male'):
        train['Sex'][i] = 1
    else:
        train['Sex'][i] = 0
```

Видим, что самыми влияющими параметрами являются: 'Pclass', 'Sex', 'Embarked' и 'Fare'. Именно по этим параметрам и будет строиться обучающая модель.

Для обучения модели будем использовать метод случайных деревьев.

```python
test = pd.read_csv("titnic_test.csv")
test.isnull().sum()
```

```python
fares = test['Fare']
sum = 0
for i in range(0, len(fares)):
    if (fares[i] >= 0):
        sum += fares[i]

average = sum/(len(fares)-1)

test['Fare'] = test['Fare'].fillna(average)
```

```python
for i in range(0, len(test['Embarked'])):
    if (test['Embarked'][i] == 'C'):
        test['Embarked'][i] = 0
    if (test['Embarked'][i] == 'Q'):
        test['Embarked'][i] = 1
    if (test['Embarked'][i] == 'S'):
        test['Embarked'][i] = 2
```

```python
for i in range(0, len(test['Sex'])):
    if (test['Sex'][i] == 'male'):
        test['Sex'][i] = 1
    else:
        test['Sex'][i] = 0
```

```python
test
```

```python
from sklearn.ensemble import RandomForestClassifier

Y = train["Survived"]

features = ["Pclass", "Sex", "Fare", "Embarked"]
X = pd.get_dummies(train[features])
X_test = pd.get_dummies(test[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, Y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('prediction2_1.csv', index=False)
```
